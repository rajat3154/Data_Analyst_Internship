```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.info();
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.value_counts()
```

```python
df['Sex'].value_counts().plot(kind='bar', title='Gender Distribution',color='blue')
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='bar',color='blue', title='Passenger Class Distribution')
plt.show()
```

```python
df['Age'].hist(bins=30,color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x=df['Fare'],color='blue')
plt.title('Fare Boxplot')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Survived']], hue='Survived')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df,color='blue')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df,color='blue')
plt.title('Survival Rate by Passenger Class')
plt.show()
```
